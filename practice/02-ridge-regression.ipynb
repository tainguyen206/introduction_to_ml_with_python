```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Ridge

X, y = mglearn.datasets.load_extended_boston()

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)

lr = LinearRegression().fit(X_train, y_train)
print(
    f'Linear:'
    f'\ntrain score: {lr.score(X_train, y_train)}'
    f'\ntest score: {lr.score(X_test, y_test)}'
    f'\nintercept: {lr.intercept_}'
    # f'\ncoef/slop: {lr.coef_}'
)

ridge = Ridge().fit(X_train, y_train)
print(
    f'Ridge alpha=1:'
    f'\ntrain score: {ridge.score(X_train, y_train)}'
    f'\ntest score: {ridge.score(X_test, y_test)}'
    f'\nintercept: {ridge.intercept_}'
    # f'\ncoef/slop: {ridge.coef_}'
)

ridge10 = Ridge(alpha=10).fit(X_train, y_train)
print(
    f'Ridge alpha=10:'
    f'\ntrain score: {ridge10.score(X_train, y_train)}'
    f'\ntest score: {ridge10.score(X_test, y_test)}'
    f'\nintercept: {ridge10.intercept_}'
    # f'\ncoef/slop: {ridge10.coef_}'
)

ridge01 = Ridge(alpha=0.1).fit(X_train, y_train)
print(
    f'Ridge alpha=0.1:'
    f'\ntrain score: {ridge01.score(X_train, y_train)}'
    f'\ntest score: {ridge01.score(X_test, y_test)}'
    f'\nintercept: {ridge01.intercept_}'
    # f'\ncoef/slop: {ridge01.coef_}'
)

```

```python
plt.plot(lr.coef_, 'o', label='linear')
plt.plot(ridge.coef_, 's', label='ridge alpha=1')
plt.plot(ridge10.coef_, '^', label='ridge alpha=10')
plt.plot(ridge01.coef_, 'v', label='ridge alpha=0.1')

plt.xlabel('features')
plt.ylabel('coef')
plt.ylim(-25, 25)
plt.legend()
```
